```python
from typing import TypedDict, List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
  name: str
  values: List[int]
  operation: str
  answer: int
```

```python
def multiply(numbers):
    result = 1
    for num in numbers:
        result *= num
    return result
```

```python
def apply_operation(state: AgentState) -> AgentState:
  if state["operation"] == "+":
    state["answer"] = f"Hi {state["name"]}! Your answer is: {sum(state["values"])}"
  elif state["operation"] == "*":
    state["answer" ] = f"Hi {state["name"]}! Your answer is: {multiply(state["values"])}"
  else:
    state["answer"] = f"Invalid operaton. Please use + or *."

  return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("operation", apply_operation)
graph.set_entry_point("operation")
graph.set_finish_point("operation")

app = graph.compile()
```

```python
results = app.invoke({"name": "YG", "values": [1,2,3,4], "operation": "*"})
```
